```python
data = {
    'Name':['Alice','Bob','Charlie','David','Eva'],
    'Age':[20.0,np.nan,22.0,np.nan,25.0],
    'Grade':[80.0,np.nan,np.nan,95.0,88.0],
    'Attendance':['Present','Absent',np.nan,'Present',np.nan]
}
df3 = pd.DataFrame(data)
df3
```

```python
df3.isnull().sum()
```

```python
df3['Grade'] = df3['Grade'].replace(to_replace=np.nan,value=df3['Grade'].mean())
```

```python
df3['Age'] = df3['Age'].replace(to_replace=np.nan,value=df3['Age'].median())
```
